```python
# imports
import numpy as np 
import pandas as pd 
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from matplotlib.figure import Figure

mpl.style.use('ggplot') # optional: for ggplot-like style
```

```python
df_survey = pd.read_csv('https://cocl.us/datascience_survey_data', index_col=[0])
df_survey
```

```python
df_survey.sort_values(by='Very interested', ascending=False, inplace=True)
```

```python
df_survey = round(df_survey / 2233, 2)
df_survey
```

```python
df_survey.T
```

```python
df_survey.dtypes
```

```python
from matplotlib.lines import Line2D   
```

```python
ax = df_survey.plot(kind='bar', figsize=(20,8), width=0.8, color=['#5cb85c', '#5bc0de', '#d9534f'])

ax.set_title("Percentage of Respondents' Interest in Data Science Areas", fontsize=16)

ax.set_xticklabels(df_survey.index.tolist(), fontsize=14)

ax.set_facecolor('white')

for p in ax.patches: 
    ax.annotate(np.round(p.get_height(), 2), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=14)


ax.get_yaxis().set_visible(False)

ax.get_xaxis().tick_bottom()

ax.legend(fontsize=14, facecolor='white')

# code to draw line at the bottom
xmin, xmax = ax.get_xaxis().get_view_interval()
ymin, ymax = ax.get_yaxis().get_view_interval()
ax.add_artist(Line2D((xmin, xmax), (ymin, ymin), color='grey', linewidth=6, alpha=0.6));
```

```python
# read in data
df_crime = pd.read_csv('https://cocl.us/sanfran_crime_dataset')
df_crime.head()
```

```python
# print shape of dataframe
print(df_crime.shape)
```

```python
# get total number of crimes by neighborhood
df_crime_n = df_crime['PdDistrict'].value_counts()
df_crime_n = df_crime_n.to_frame()
```

```python
# reset index
df_crime_n.reset_index(inplace=True)
```

```python
# rename columns
df_crime_n.columns = ['Neighborhood', 'Count']
df_crime_n
```

```python
import json
import folium
import requests
```

```python
# load SF GeoJSON
r = requests.get('https://cocl.us/sanfran_geojson')
sf_geo = r.json()
print(sf_geo)
```

```python
# create df with neighborhood locations
nb_location = df_crime.groupby('PdDistrict').mean()
```

```python
# San Francisco latitude and longitude values
latitude = 37.77
longitude = -122.42

# create map and display it
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# add markers for neighborhoods
labels = list(nb_location.index)

latitudes = list(nb_location.Y)
longitudes = list(nb_location.X)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sanfran_map) 

# display the map of San Francisco
sanfran_map
```

```python
sanfran_map.choropleth(
    geo_data=sf_geo,
    data=df_crime_n,
    columns=['Neighborhood', 'Count'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Rate in San Francisco',
    reset=True
)

sanfran_map
```
